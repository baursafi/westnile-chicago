```python
# epicenter being set at Trap T160
# calculating radial coordinates for T160
df_train[df_train['Trap'] == 'T160'][['Latitude','Longitude']].head(1) # getting lat/long for epicenter
```

```python
epicenter_lat = 41.903002
epicenter_long = -87.688267
```

```python
epicenter_rho = np.sqrt(epicenter_lat**2 + epicenter_long**2)
epicenter_phi = np.arctan2(epicenter_long, epicenter_lat)
```

```python
print epicenter_rho
print epicenter_phi
```

```python
df_train['rho'] = np.sqrt((df_train['Latitude'] - epicenter_lat)**2 + (df_train['Longitude'] - epicenter_long)**2)
```

```python
df_train['phi'] = np.arctan2(df_train['Longitude'], df_train['Latitude'])
```

```python
# rho = np.sqrt(x**2 + y**2)
# phi = np.arctan2(y, x)
df_train['PC'] = list(zip(df_train['rho'], df_train['phi']))
```
